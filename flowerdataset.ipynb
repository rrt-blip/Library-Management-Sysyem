```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
#reading the dataset
iris = pd.read_csv('/content/iris.csv')
print(iris.head())
#creating the test split model
X_train, X_test, y_train, y_test = train_test_split(iris.drop('Class', axis=1), iris['Class'], test_size=0.3, random_state=42)
print()

#Decision tree classifier
tree = DecisionTreeClassifier(min_samples_split=5, min_samples_leaf=3, max_depth=3, random_state = 42)
tree.fit(X_train, y_train)
#Prediction vs Testing
y_pred = tree.predict(X_test)
print("Predicion: \n")
print(y_pred)
print("\nTest: \n")
print(y_test)

print("\n\n")

#Confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
classes = ['setosa', 'versicolor', 'virginica']
cm_df = pd.DataFrame(conf_matrix, index=classes, columns=classes)
print("\nConfustion Matrix\n")
print(cm_df)
print("\n\n")
print("Classification report\n")
print(classification_report(y_test, y_pred))
print("\n\n")
print("----------------------------------------------------------------------")
print("Random Forest")

rf = RandomForestClassifier(n_estimators = 100, random_state = 42)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
print("Predicion: \n")
print(y_pred_rf)
print("\nTest: \n")
print(y_test)
print("\nClassification report")
print(classification_report(y_test, y_pred_rf))

conf_matrix_rf = confusion_matrix(y_test, y_pred_rf)
cm_df_rf = pd.DataFrame(conf_matrix_rf, index=classes, columns=classes)
print("\nConfustion Matrix\n")
print(cm_df_rf)

print("\n----------------------------------------------------------------------\n")
print("SVM")

svm = SVC(C=10.0)
svm.fit(X_train, y_train)
y_pred_svm = svm.predict(X_test)
print("Predicion: \n")
print(y_pred_svm)
print("\nTest: \n")
print(y_test)

conf_matrix_svm = confusion_matrix(y_test, y_pred_svm)
cm_df_svm = pd.DataFrame(conf_matrix_svm, index=classes, columns=classes)
print("\nConfustion Matrix\n")
print(cm_df_svm)

print("\nClassification report")
print(classification_report(y_test, y_pred_svm))

```
